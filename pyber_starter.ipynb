```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
ride_data_complete_df = pd.merge(city_data_df, ride_data_df, how="left", on=["city", "city"])

# Display the data table for preview
ride_data_complete_df.head()


# Create a group based on the values in the 'type' column
city_name_group = ride_data_complete_df.groupby('type')

# Count how many times each maker appears in our group
count_city_type = city_name_group ['type'].count()

count_city_type
```

```python
# Create a group based on the values in the 'city' column
city_name_group = ride_data_complete_df.groupby('city')

# Count how many times each maker appears in our group
count_city = city_name_group ['city'].count()

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# The max x value for chart is 100
x_limit = 100

# List of values 0-100, increasing by 1
x_axis = np.arange(0, x_limit, 1)

#Create array
xdata = ride_data_complete_df['city']
ydata = ride_data_complete_df['fare']

#plt.scatter(x=ride_data_complete_df ['city'], y=ride_data_complete_df ['fare'])
plt.scatter(xdata, ydata, marker="o", facecolors="yellow", edgecolors="black", s=x_axis, alpha=0.5)


# Incorporate the other graph properties
plt.title('Average Fare ($) Per City')
plt.xlabel('Total Number of Rides Per City')
plt.ylabel('Average Fare ($)')

# Show plot
plt.show()


# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```
